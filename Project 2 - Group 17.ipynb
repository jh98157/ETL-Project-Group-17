```python
# Dataframe 1 
for x in range(1000):
    movie_title=str(df1.at[x,"movie_title"])
    response=requests.get(f"{url} + {movie_title}").json()
    try:
        imdb=(response['imdbRating'])
        runtime=(response['Runtime'])
        rotten_tomatoes=(response["Ratings"][1]["Value"])
        image=(response['Poster'])
    except(KeyError,IndexError):
            print(f"Movie{x}: No data Found")
            
    df1.at[x,'imdb_rating']=imdb
    df1.at[x,'Runtime']=runtime
    df1.at[x,'Rotten Tomatoes Score']=rotten_tomatoes
    df1.at[x,'Poster']=image
```
